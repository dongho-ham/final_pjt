```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
battery_ch = pd.read_csv('B0005_charge.csv')
battery_dis = pd.read_csv("B0005_discharge.csv")
```

# 충전 데이터셋

```python
battery_ch.head()
```

```python
battery_ch.info()
```

```python
battery_ch['type'].value_counts()
```

```python
col_plot = ['Voltage_measured', 'Current_measured', 'Temperature_measured', 'Current_charge', 'Voltage_charge', 'ambient_temperature']

num_plots = len(col_plot)
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

axes = axes.flatten()

for i, col in enumerate(col_plot):
    sns.boxplot(data=battery_ch, y=col, ax=axes[i], color='skyblue') # 모든 박스플롯 색상을 'skyblue'로 통일
    axes[i].set_title(f'{col} Distribution') # 컬럼 이름을 기반으로 제목 설정
    axes[i].set_ylabel(col) # 컬럼 이름을 기반으로 y축 레이블 설정
    axes[i].tick_params(axis='x', bottom=False)
    axes[i].set_xlabel('')

plt.tight_layout()

plt.show()
```

```python
col_plot = ['Voltage_measured', 'Current_measured', 'Temperature_measured', 'Current_charge', 'Voltage_charge']

num_plots = len(col_plot)
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

axes = axes.flatten()

for i, col in enumerate(col_plot):
    sns.lineplot(data=battery_ch, x=battery_ch.index, y=col, ax=axes[i], color='skyblue') # 모든 박스플롯 색상을 'skyblue'로 통일
    axes[i].set_title(f'{col} Distribution') # 컬럼 이름을 기반으로 제목 설정
    axes[i].set_ylabel(col) # 컬럼 이름을 기반으로 y축 레이블 설정
    axes[i].tick_params(axis='x', bottom=False)
    axes[i].set_xlabel('')

plt.tight_layout()

plt.show()
```

```python
print('Current_measured min: ', battery_ch['Current_measured'].min())
print('Current_measured max: ', battery_ch['Current_measured'].max())
```

```python
print('Temperature_measured min: ', battery_ch['Temperature_measured'].min())
print('Temperature_measured max: ', battery_ch['Temperature_measured'].max())
```

```python
print('Voltage_charge min: ', battery_ch['Voltage_charge'].min())
print('Voltage_charge max: ', battery_ch['Voltage_charge'].max())
```
